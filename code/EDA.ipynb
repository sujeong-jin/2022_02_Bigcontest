```python
def convert_time(x):
    if np.isnan(x):
        return x
    if np.log10(x)>7:
        x = x // 100
    x = pd.to_datetime(str(int(x)), format='%Y%m')
    return x
```

```python
foo = user_spec["company_enter_month"]
foo = foo.map(convert_time)
user_spec["company_enter_month"] = foo
```

적절히 `dtype`을 변경하고, 시간을 나타내는 변수들에 대해서 적당히 `datetime64`로 변경한다.

`user_spec2`에 있는 생년과 성별 처리

```python
user_info = user_spec2.groupby('user_id')[['birth_year', 'gender']].mean()
user_info[(user_info['gender'] == 1.0) & (user_info['gender'] == 0.0)] # 성별이 달라지거나 생년이 달라지는 사람 없음!
```
